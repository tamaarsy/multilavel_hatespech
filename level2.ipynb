```python
df_1 = pd.read_csv('data_cleaned_new.csv')
df_1.drop(['Unnamed: 0'],axis=1, inplace=True)
df_1.tail()
```

```python
x_train, x_val, y_train, y_val = train_test_split(df_1['str_filtered'], df_1['HS'], test_size=0.3, random_state=42)
```

```python
#tokenizer data
vocab_size = 1000
oov_token = "<OOV>"
tokenizer = Tokenizer(num_words = vocab_size, oov_token=oov_token)
tokenizer.fit_on_texts(x_train)
```

```python
word_index = tokenizer.word_index
```

```python
#sequence data final
x_train_sequences = tokenizer.texts_to_sequences(x_train)
x_val_sequences = tokenizer.texts_to_sequences(x_val)
```

```python
#padding_sequences(diurutkan)
max_length = 400
padding_type = "post"
truncation_type = "post"
```

```python
x_train_padded = pad_sequences(x_train_sequences, maxlen=max_length, padding=padding_type,
                               truncating = truncation_type)

x_val_padded = pad_sequences(x_val_sequences, maxlen=max_length, padding=padding_type,
                               truncating = truncation_type)
```

```python
train_tensor = tf.data.Dataset.from_tensor_slices((x_train_padded, y_train))
val_tensor = tf.data.Dataset.from_tensor_slices((x_val_padded, y_val))
```
